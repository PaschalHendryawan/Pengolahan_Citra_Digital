```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = Image.open("citraa.jpg")

# Convert the image to a NumPy array
image_array = np.array(image)

# Separate the R, G, and B channels
red_channel = image_array[:, :, 0]
green_channel = image_array[:, :, 1]
blue_channel = image_array[:, :, 2]

# Create an empty image for each channel
red_image = Image.fromarray(red_channel, 'L')
green_image = Image.fromarray(green_channel, 'L')
blue_image = Image.fromarray(blue_channel, 'L')

# Convert the original image to grayscale
gray_image = image.convert('L')

# Display the images
plt.figure(figsize=(15, 5))

# Display the original image
plt.subplot(151)
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')

# Display the R, G, and B channels
plt.subplot(152)
plt.title("Red Channel")
plt.imshow(red_image, cmap='Reds')
plt.axis('off')

plt.subplot(153)
plt.title("Green Channel")
plt.imshow(green_image, cmap='Greens')
plt.axis('off')

plt.subplot(154)
plt.title("Blue Channel")
plt.imshow(blue_image, cmap='Blues')
plt.axis('off')

# Display the grayscale image
plt.subplot(155)
plt.title("Grayscale Image")
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.show()

# Optionally, you can modify the grayscale image or save it
# Example: Save the grayscale image
gray_image.save("grayscale_image.jpg")
```
